## Nodes(Units)

A fundamental component of deep learning is the artificial neuron, which is inspired by the
biological neuron found in the human brain. These artificial neurons, also known as `nodes`
or `units`, are organized into layers within a neural network. The `input layer` receives the raw
data, while the `hidden layers` progressively transform and abstract the information. The
`output layer` produces the final prediction or decision based on the learned features

![HL](https://miltonfmr.com/wp-content/uploads/2021/04/artificial-neural-network.png)

## Weights

Each connection between neurons in adjacent layers is associated with a `weight`, which
determines the strength of the connection. During training, the neural network adjusts these
weights through a process called backpropagation, which minimizes a specific loss functionby iteratively updating the weights to reduce prediction errors. This training process allows the network to learn from labeled data and improve its performance.

Each neuron in a neural network takes multiple input values, processes them using
weighted connections, and produces an output value. The weighted connections represent
the strength of the influence of each input on the neuron's output. These weights are
learned during the training process and are adjusted to optimize the network's performance
on a specific task.

## Activation Functions

`Neurons` and `activation functions` are fundamental components of artificial neural networks,
forming the building blocks of deep learning models. These components play a crucial role
in processing and transforming information within a neural network. Neurons are inspired by
biological neurons found in the human brain, which transmit signals and process information
through electrical and chemical signals. In artificial neural networks, neurons are
mathematical units that receive input, perform computations, and produce output.

The output of a neuron is determined by an `activation function`, which is a mathematical
function that introduces non-linearity into the network. Activation functions are essential
because they allow neural networks to model complex relationships and capture non-linear
patterns in data. Without activation functions, neural networks would reduce to linear

## Fully Connected Neural Network (Dense Layer)

A `Fully Connected (FC) layer`, also known as a `dense layer`, is a type of layer used in artificial neural networks where each neuron or node from the previous layer is connected to each neuron of the current layer. It's called “fully connected” because of this `complete linkage`.

![FC](https://media.licdn.com/dms/image/D5634AQEVPkms69HDRA/ugc-proxy-shrink_1280_800/0/1669827603355?e=2147483647&v=beta&t=JA-mfObMPxLvd7296rqMdMxvQkpvrRBSNCKIUzKQcmw)

## Forward & Backward Propagation

`Forward propagation` is the process in a neural network where the input data is passed through the network's layers to generate an output. It involves the following steps:
1. `Input Layer`: The input data is fed into the input layer of the neural network.
2. `Hidden Layers`: The input data is processed through one or more hidden layers. Each neuron in a hidden layer receives inputs from the previous layer, applies an activation function to the weighted sum of these inputs, and passes the result to the next layer.
3. `Output Layer`: The processed data moves through the output layer, where the final output of the network is generated.
The output layer typically applies an activation function suitable for the task.
4. `Prediction`: The final output of the network is the prediction or classification result for the input data.
- Forward propagation is essential for making predictions in neural networks.

It calculates the output of the network for a given input based on the current values of the weights and biases. The output is then compared to the actual target value to calculate the loss, which is used to update the weights and biases during the training process.

![FP](https://anatomiesofintelligence.github.io/img/algorithms/forward-back-propagation-neuralnetworks.gif)

Backpropagation is a crucial technique in deep learning for training neural networks. It involves two main processes: `forward propagation`, where input data is processed through the network to produce an output, and `backward propagation`, where errors are propagated backward to adjust the network's weights. This adjustment minimizes the cost function, improving the model's accuracy over time. The process leverages the chain rule of calculus to compute gradients efficiently. Different types of backpropagation algorithms, including static and recurrent backpropagation, are each suited to other network architectures. Backpropagation's efficiency, speed, versatility, and simplicity make it a widely used method in various neural network applications, including TensorFlow and PyTorch frameworks.

## Loss function 

`Loss functions` are one of the most important aspects of neural networks, as they (along with the optimization functions) are directly responsible for fitting the model to the given training data.

A `loss function` is a function that compares the target and predicted output values; measures how well the neural network models the training data. When training, we aim to minimize this loss between the predicted and target outputs.

$$Loss = \sum\limits _{i=1} ^{n} (y_{i} - \hat{y_{i}})^2$$

![LF](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSZwrRr0war2OkNyhNcwRCyMefa5uSLktZ52w&s)

We can use optimization, if the value of Loss Function is high:

$$w_{new} = \eta \frac{∂L}{∂w_{old}}+ w_{old} $$

OR:

$$w_{new} = -\eta \frac{∂L}{∂w_{old}}+ w_{old} $$

### Chain Rule in Backward Propagation

Backward Propagation is Starting from the output layer, the gradient of the loss function with respect to the output is computed. The **chain rule** is applied iteratively to compute the gradient of the loss function with respect to each layer's parameters and inputs.

$$Loss = (y_{i} - \hat{y_{i}})^2 + (y_{i} - \hat{y_{i}})^2 + (y_{i} - \hat{y_{i}})^2 + ...$$ 

![CR](https://media.geeksforgeeks.org/wp-content/uploads/20240217152156/Frame-13.png)


$$\frac{∂L}{∂w_{11}^{3}} = \frac{∂L}{∂O_{31}} * \frac{∂O_{31}}{∂w_{11}^{3}}$$

One step back:

$$\frac{∂L}{∂w_{11}^{2}} = \frac{∂L}{∂w_{11}^{3}} * \frac{∂w_{11}^{3}}{∂O_{21}} * \frac{∂O_{21}}{∂w_{11}^{2}}$$

And so on for the rest of the steps...

## Weighted Sum

This `weighted sum` is considered the pre-activation output from the node. This output will be passed to a non-linear activation function before passing the final output as input to the following layer via the connected weights.

$$Weighted  sum = \sum\limits _{i=1} ^{n} x_{i}w_{i}$$

`Weighted sum` is the sum of any number multiplied by its row and column index starting at 1 and the sum is addition of all values of a `matrix`.
$$y = \sum\limits _{i=1} ^{n} x_{i}w_{i} + b$$

- For example:

$$\left[\begin{matrix} y_{1} & y_{2} & \cdots & y_{m} \end{matrix}\right]_{1*m} = \left[\begin{matrix} x_{1} & x_{2} & \cdots & x_{n} \end{matrix}\right]_{1*n} \left[\begin{matrix} w_{11} & w_{12} & \cdots & w_{nm} \\ w_{21} & w_{22} & \cdots & w_{nm} \\  \vdots & \vdots & \ddots & \vdots \\ w_{31} & w_{32} & \cdots & w_{nm} \end{matrix}\right]_{n*m} + \left[\begin{matrix} b_{1} & b_{2} & \cdots & b_{m} \end{matrix}\right]_{1*m} $$

- **If W is a `n*m` Matrix, X is a `1*n` Matrix, and target(Y) is a `1*m` Matrix, B(bias) will be a `1*m` Matrix.**

### Activation Functions in Deep Learning

#### Sigmoid/ Logistic

One of the most commonly used activation functions is the **sigmoid function**. The sigmoid
function maps the weighted sum of inputs to a value `between 0 and 1`, making it suitable for
`binary classification` problems where the output represents the probability of a certain class.
However, sigmoid functions `suffer from the vanishing gradient problem`, which can slow
down training in deep networks.

$$f(x) = \frac {1}{1 + e^{-x}}$$

- 3 `disadvantages` of Sigmoid function:
1. **Gradient Vanishing**: The slope is zero in the points of 0 and 1
2. **Exponential**: A dumb number to the power of a big number makes training harder and longer
3. The sigmoid function is **symmetric** around its midpoint at x=0.5.

- 3 `advantages` of Sigmoid function:
1. It's suitable for **binary classification**.
2. **Normalize numbers**
3. Because it is between 0 and 1, it can be a **probability model**.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Convert x to a symbol
x = sp.symbols('x')
print(type(x))
x
```

```python
# Definition of sigmoid
sigmoid = 1 / (1 + sp.exp(-x))
sigmoid
```

```python
# Definition of derivative of sigmoid
sigmoid_derivative = sp.diff(sigmoid, x)
sigmoid_derivative
```

```python
# Convert a symbolic variable to a non-symbolic variable
sigmoid_function = sp.lambdify(x, sigmoid, 'numpy')
sigmoid_derivative_function = sp.lambdify(x, sigmoid_derivative, 'numpy')
```

```python
# Calculate a point with its derivative
x0 = 2.5
y0 = sigmoid_function(x0)
y0d = sigmoid_derivative_function(x0)
print(f'[{x0}, {y0}]')
print(f'[{x0}, {y0d}]')
```

```python
# Definition of x and y
x = np.linspace(-10, 10, 400)
```

```python
y = sigmoid_function(x)
```

```python
# Show plots
plt.figure(figsize=(6, 3))
plt.plot(x, y, label='Sigmoid', color='purple')
plt.scatter(x[200], y[200], marker='x', c='k')
plt.scatter(x0, y0, marker='x', c='k')
plt.grid(True)
plt.legend()
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
yd = sigmoid_derivative_function(x)
plt.figure(figsize=(6, 3))
plt.plot(x, yd, label='Sigmoid Derivative', color='red', linestyle="--")
plt.scatter(x[200], yd[200], marker='x', c='k')
plt.scatter(x0, y0d, marker='x', c='k')
plt.title('Sigmoid Derivative')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x, y, label='Sigmoid', color='purple')
plt.plot(x, yd, label='Sigmoid Derivative', color='red', linestyle="--")
plt.legend()
plt.title('Sigmoid and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

#### Hyperbolic Tangent (tanh)

Another activation function called the **hyperbolic tangent (tanh)** function is similar to the
sigmoid function but maps inputs to values between -1 and 1. Tanh is often used in
recurrent neural networks (RNNs) and can be advantageous when data is centered around zero.

$$tanh(x) = \frac{e^{x} - e^{-x}} {e^{x} + e^{-x}}$$

- **It is like sigmoid with the difference that hyperbolic tangent is between 1 and -1 and the problem of being symmetrical is solved.**

```python
# Convert x to a symbol
x = sp.symbols('x')
tanh = sp.tanh(x)
tanh
```

```python
# Definition of derivative of tanh
tanh_der = sp.diff(tanh, x)
tanh_der
```

```python
# Convert a symbolic variable to a non-symbolic variable
tanh_func = sp.lambdify(x, tanh, 'numpy')
tanh_der_func = sp.lambdify(x, tanh_der, 'numpy')
```

```python
# Definition of x and y
x_vals = np.linspace(-10, 10, 400)
y_vals = tanh_func(x_vals)
yd_vals = tanh_der_func(x_vals)
```

```python
# Show plots
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_vals, label='tanh', color='purple')
plt.grid(True)
plt.legend()
plt.title('tanh')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_vals, yd_vals, label='d[tanh]/dx', color='red', linestyle="--")
plt.title('tanh Derivative')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_vals, label='tanh', color='purple')
plt.plot(x_vals, yd_vals, label='d[tanh]/dx', color='red', linestyle='--')
plt.title('tanh and its Derivative')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Rectified Linear Unit (RelU)

To `address` the `vanishing gradient problem`, the `rectified linear unit (ReLU) activation
function` was introduced. ReLU is defined as the maximum of zero and the weighted sum of
inputs. ReLU is computationally efficient and allows the network to learn faster and capture
complex patterns. However, ReLU has its own challenges, such as the "dying ReLU"
problem, where neurons can become inactive during training.

- **the slope of this function is zero at negative points**.

$$f(x) = \begin{cases} 0 &  x<0 \\ x & x>=0 \end{cases}$$

```python
# Convert x to a symbol and write piecewise function of RelU
relu = sp.Piecewise(
     (0, x < 0),
     (x, x >= 0)
 )
relu
```

```python
# Definition of derivative of RelU
relu_der = sp.diff(relu, x)
relu_der
```

```python
# Convert a symbolic variable to a non-symbolic variable
relu_fun = sp.lambdify(x, relu, 'numpy')
relu_der_fun = sp.lambdify(x, relu_der, 'numpy')
```

```python
# Definition of x and y
y_relu = relu_fun(x_vals)
y_relu_der = relu_der_fun(x_vals)
```

```python
# Show plots
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_relu, label='RelU', color='purple')
plt.grid(True)
plt.legend()
plt.title('RelU')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_relu_der, label='d[ReLU]/dx', color='red', linestyle="--")
plt.title('RelU Derivative')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_relu, color='purple', label='ReLU')
plt.plot(x_vals, y_relu_der, color='red', linestyle='--', label='d[ReLU]/dx')
plt.title('RelU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

#### Leaky RelU

`Leaky ReLU` is a variant of the ReLU activation function that addresses the dying ReLU problem by allowing a small, non-zero gradient for negative input values. This ensures that even neurons that have learned poorly still contribute to the training process.

- **Unlike Relu, the slope of this function is not zero at negative points and is equal to alpha value**.

$$f(x) = \begin{cases} \alpha x &  x<0 \\ x & x>=0 \end{cases}$$

```python
# Convert x to a symbol and write piecewise function of Leaky RelU
x = sp.symbols('x')
alpha = 0.01
leaky_relu = sp.Piecewise((alpha * x, x < 0),
                         (x, x >= 0))
leaky_relu
```

```python
# Definition of derivative of Leaky RelU
leaky_relu_derivative = sp.diff(leaky_relu, x)
leaky_relu_derivative
```

```python
# Convert a symbolic variable to a non-symbolic variable
leaky_relu_func = sp.lambdify(x, leaky_relu, 'numpy')
leaky_relu_derivative_func = sp.lambdify(x, leaky_relu_derivative,'numpy')
```

```python
# Definition of x and y
x_lrelu = np.linspace(-10, 10, 400)
y_lrelu = leaky_relu_func(x_lrelu)
y_dlrelu = leaky_relu_derivative_func(x_lrelu)
```

```python
# Show plots
plt.figure(figsize=(6, 3))
plt.plot(x_lrelu, y_lrelu, label='Leaky RelU', color='purple')
plt.grid(True)
plt.legend()
plt.title('Leaky RelU')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_lrelu, y_dlrelu, label='d[Leaky ReLU]/dx', color='red', linestyle="--")
plt.title('Leaky RelU Derivative')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_lrelu, y_lrelu, color='purple', label=f'Leaky RelU (α={alpha})')
plt.plot(x_lrelu, y_dlrelu, color='red', linestyle='--', label='Leaky RelU Derivative')
plt.title('Leaky RelU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exponential Linear Unit (ELU)

`Exponential Linear Unit` or its widely known name `ELU` is a function that tend to converge cost to zero faster and produce more accurate results. Different to other activation functions, ELU has a extra alpha constant which should be positive number.

ELU is very similiar to RELU except negative inputs. They are both in identity function form for non-negative inputs. On the other hand, ELU becomes smooth slowly until its output equal to -α whereas RELU sharply smoothes.

- **In the negative part, the slope of the graph is exponential**.

$$f(x) = \begin{cases} \alpha (e^{x}-1) &  x<0 \\ x & x>=0 \end{cases}$$

```python
# Convert x to a symbol and write piecewise function of ELU
x = sp.symbols('x')
alpha = 1
elu = sp.Piecewise((alpha * (sp.exp(x) - 1), x < 0), (x, x >= 0))
elu
```

```python
# Definition of derivative of ELU
elu_derivative = sp.diff(elu, x)
elu_derivative
```

```python
# Convert a symbolic variable to a non-symbolic variable
elu_func = sp.lambdify(x, elu, 'numpy')
elu_derivative_func = sp.lambdify(x, elu_derivative, 'numpy')
```

```python
# Definition of x and y
x_elu = np.linspace(-10, 10, 400)
y_elu = elu_func(x_elu)
y_delu = elu_derivative_func(x_elu)
```

```python
# Show plots
plt.figure(figsize=(6, 3))
plt.plot(x_elu, y_elu, label='ELU', color='purple')
plt.grid(True)
plt.legend()
plt.title('ELU')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_elu, y_delu, label='d[ELU]/dx', color='red', linestyle="--")
plt.title('ELU Derivative')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(x_elu, y_elu, color='purple', label=f'ELU (α={alpha})')
plt.plot(x_elu, y_delu, color='red', linestyle='--', label='ELU Derivative')
plt.title('ELU and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
